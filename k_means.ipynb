```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from matplotlib.colors import ListedColormap
from pprint import pprint
import matplotlib.pyplot as mtp    
```

```python
file_path = '/Users/chiaradraghini/Documents/mrn_project/data/k-means/dataset/'
dataframe = pd.read_csv(file_path+'weekdays.csv') 
dataframe
```

```python
scaler = StandardScaler()
cells = dataframe["ECELL_ID"].to_list()
cells_df = pd.DataFrame(cells)
cells_df.columns = ["ECELL_ID"]
df_filtered = dataframe.drop(["ECELL_ID"], axis=1)
```

```python
scaled_array = scaler.fit_transform(df_filtered)
scaled_dataframe = pd.DataFrame(scaled_array)
print(scaled_dataframe)
```

```python
kmeans_model = KMeans(n_clusters = 3)
kmeans_model.fit(scaled_dataframe)
centroids = kmeans_model.cluster_centers_
print(centroids)
```

```python
np.savetxt("data/k-means/results/centroids_kmeans.txt", centroids, fmt="%s")
```

```python
#labels corresponding to those clusters
kmeans_model.labels_
```

```python
cells_df["Cluster"] = kmeans_model.labels_
cells_df.to_csv("/Users/chiaradraghini/Documents/mrn_project/data/k-means/results/Clusters_weekdays.csv")
```
