```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from matplotlib.colors import ListedColormap
from pprint import pprint
import matplotlib.pyplot as mtp  
from scipy import stats
```

```python
file_path = '/Users/chiaradraghini/Documents/mrn_project/data/k-means/dataset/'
dataframe = pd.read_csv(file_path+'weekdays.csv') 
dataframe
```

```python
scaler = StandardScaler()
cells = dataframe["ECELL_ID"].to_list()
cells_df = pd.DataFrame(cells)
df_filtered = dataframe.drop(["ECELL_ID"], axis=1)
cols_to_keep = ["Friday at 0", "Friday at 1", "Friday at 2", "Friday at 3", "Friday at 4",
                "Friday at 5", "Friday at 6", "Friday at 7", "Friday at 8", "Friday at 9", 
                "Friday at 10", "Friday at 11", "Friday at 12", "Friday at 13", "Friday at 14", 
                "Friday at 15", "Friday at 16", "Friday at 17", "Friday at 18", "Friday at 19", 
                "Friday at 20", "Friday at 21", "Friday at 22", "Friday at 23", 
                "Saturday at 0", "Saturday at 1", "Saturday at 2", "Saturday at 3", "Saturday at 4",
                "Saturday at 5", "Saturday at 6", "Saturday at 7", "Saturday at 8", "Saturday at 9", 
                "Saturday at 10", "Saturday at 11", "Saturday at 12", "Saturday at 13", "Saturday at 14", 
                "Saturday at 15", "Saturday at 16", "Saturday at 17", "Saturday at 18", "Saturday at 19", 
                "Saturday at 20", "Saturday at 21", "Saturday at 22", "Saturday at 23",
                "Sunday at 0", "Sunday at 1", "Sunday at 2", "Sunday at 3", "Sunday at 4",
                "Sunday at 5", "Sunday at 6", "Sunday at 7", "Sunday at 8", "Sunday at 9", 
                "Sunday at 10", "Sunday at 11", "Sunday at 12", "Sunday at 13", "Sunday at 14", 
                "Sunday at 15", "Sunday at 16", "Sunday at 17", "Sunday at 18", "Sunday at 19", 
                "Sunday at 20", "Sunday at 21", "Sunday at 22", "Sunday at 23"]
df_filtered_by_days = df_filtered.loc[:,cols_to_keep]
```

```python
scaled_array = scaler.fit_transform(df_filtered_by_days)
scaled_dataframe = pd.DataFrame(scaled_array)
print(scaled_dataframe)
```

```python
kmeans_model = KMeans(n_clusters = 3)
kmeans_model.fit(scaled_dataframe)
centroids = kmeans_model.cluster_centers_
print(centroids)
```

```python
np.savetxt("data/k-means/results/centroids_kmeans_weekend.txt", centroids, fmt="%s")
```
