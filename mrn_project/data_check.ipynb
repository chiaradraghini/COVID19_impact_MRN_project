```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from matplotlib.colors import ListedColormap
from pprint import pprint
```

```python
file_path = '/Users/chiaradraghini/Documents/mrn_project/data/original_data/'

data_january = pd.read_csv(file_path+'Milano_800_January_MRN.csv') # Here January is selected
data_january.drop(['Unnamed: 0'], axis=1, inplace=True)
data_january['Date'] = pd.to_datetime(data_january['Date'])

data_february = pd.read_csv(file_path+'Milano_800_February_MRN.csv') # Here February is selected
data_february.drop(['Unnamed: 0'], axis=1, inplace=True)
data_february['Date'] = pd.to_datetime(data_february['Date'])

data_march = pd.read_csv(file_path+'Milano_800_March_MRN.csv') # Here March is selected
data_march.drop(['Unnamed: 0'], axis=1, inplace=True)
data_march['Date'] = pd.to_datetime(data_march['Date'])
```

```python
#checking number of different cells and which these are in January
array_cells_jan = []
counter_jan = 0
cell_id_jan = data_january["ECELL_ID"].to_list()
for i in range(len(cell_id_jan)): 
    if not (cell_id_jan[i] in array_cells_jan):
        array_cells_jan.append(cell_id_jan[i])
        counter_jan+=1
print(counter_jan)
array_cells_jan
```

```python
array_cells_feb = []
counter_feb = 0
cell_id_feb = data_february["ECELL_ID"].to_list()
for i in range(len(cell_id_feb)): 
    if not (cell_id_feb[i] in array_cells_feb):
        array_cells_feb.append(cell_id_feb[i])
        counter_feb += 1
print(counter_feb)
array_cells_feb
```
