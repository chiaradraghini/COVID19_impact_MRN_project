```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from matplotlib.colors import ListedColormap
from datetime import datetime
import time
```

***************
DATA PROCESSING PHASE 1
***************

```python
file_path = '/Users/chiaradraghini/Documents/mrn_project/data/'

data_february = pd.read_csv(file_path+'original_data/Milano_800_February_MRN.csv') 
data_february['Date'] = pd.to_datetime(data_february['Date']) 
```

```python
Mondays = pd.DataFrame()
Tuesdays = pd.DataFrame()
Wednesdays = pd.DataFrame()
Thursdays = pd.DataFrame()
Fridays = pd.DataFrame()
Saturdays = pd.DataFrame()
Sundays = pd.DataFrame()

#first days of the week in February as starting point
Mon = 3
Tue = 4
Wed = 5
Thu = 6
Fri = 7
Sat = 1
Sun = 2

#creating new csv files for each day of the week with the corresponding hourly data 
while(Mon<28):
    monday = data_february[data_february['Date'] < pd.Timestamp(year=2020, month=2, day=(Mon+1), hour=0, minute=0)]
    monday = monday[monday['Date'] > pd.Timestamp(year=2020, month=2, day=(Mon-1), hour=23, minute=59)]
    Mon = Mon + 7
    Mondays = Mondays.append(monday)
Mondays.to_csv(r'/Users/chiaradraghini/Documents/mrn_project/data/weekdays/february/Mondays_feb.csv', index=False)

while(Tue<26):
    tuesday = data_february[data_february['Date'] < pd.Timestamp(year=2020, month=2, day=(Tue+1), hour=0, minute=0)]
    tuesday = tuesday[tuesday['Date'] > pd.Timestamp(year=2020, month=2, day=(Tue-1), hour=23, minute=59)]
    Tue = Tue + 7
    Tuesdays = Tuesdays.append(tuesday)
Tuesdays.to_csv(r'/Users/chiaradraghini/Documents/mrn_project/data/weekdays/february/Tuesdays_feb.csv', index=False)

while(Wed<27):
    wednesday = data_february[data_february['Date'] < pd.Timestamp(year=2020, month=2, day=(Wed+1), hour=0, minute=0)]
    wednesday = wednesday[wednesday['Date'] > pd.Timestamp(year=2020, month=2, day=(Wed-1), hour=23, minute=59)]
    Wed = Wed + 7
    Wednesdays = Wednesdays.append(wednesday)
Wednesdays.to_csv(r'/Users/chiaradraghini/Documents/mrn_project/data/weekdays/february/Wednesdays_feb.csv', index=False)

while(Thu<28):
    thursday = data_february[data_february['Date'] < pd.Timestamp(year=2020, month=2, day=(Thu+1), hour=0, minute=0)]
    thursday = thursday[thursday['Date'] > pd.Timestamp(year=2020, month=2, day=(Thu-1), hour=23, minute=59)]
    Thu = Thu + 7
    Thursdays = Thursdays.append(thursday)
Thursdays.to_csv(r'/Users/chiaradraghini/Documents/mrn_project/data/weekdays/february/Thursdays_feb.csv', index=False)

while(Fri<30):
    friday = data_february[data_february['Date'] < pd.Timestamp(year=2020, month=2, day=(Fri+1), hour=0, minute=0)]
    friday = friday[friday['Date'] > pd.Timestamp(year=2020, month=2, day=(Fri-1), hour=23, minute=59)]
    Fri = Fri + 7
    Fridays = Fridays.append(friday)
Fridays.to_csv(r'/Users/chiaradraghini/Documents/mrn_project/data/weekdays/february/Fridays_feb.csv', index=False)

while(Sat<31):
    saturday = data_february[data_february['Date'] <= pd.Timestamp(year=2020, month=2, day=(Sat), hour=23, minute=59)]
    saturday = saturday[saturday['Date'] >= pd.Timestamp(year=2020, month=2, day=(Sat), hour=0, minute=0)]
    Sat = Sat + 7
    Saturdays = Saturdays.append(saturday)
Saturdays.to_csv(r'/Users/chiaradraghini/Documents/mrn_project/data/weekdays/february/Saturdays_feb.csv', index=False)

while(Sun<30):
    sunday = data_february[data_february['Date'] < pd.Timestamp(year=2020, month=2, day=(Sun+1), hour=0, minute=0)]
    sunday = sunday[sunday['Date'] > pd.Timestamp(year=2020, month=2, day=(Sun-1), hour=23, minute=59)]
    Sun = Sun + 7
    Sundays = Sundays.append(sunday)
Sundays.to_csv(r'/Users/chiaradraghini/Documents/mrn_project/data/weekdays/february/Sundays_feb.csv', index=False)
```

***************
DATA PROCESSING PHASE 2
***************

```python
cols_to_keep = ['ECELL_ID', 'Date', 'DL_VOL', 'UL_VOL', 'CS_SR', 'InterF_Hout_SR']

mon_february = pd.read_csv(file_path+'weekdays/february/Mondays_feb.csv')
mon_february['Date'] = pd.to_datetime(mon_february['Date'])
mon_february_DL = mon_february.loc[:,cols_to_keep]
mon_february_index = mon_february_DL.set_index("Date")

'''*****************************************************************************************'''
                                             
tue_february = pd.read_csv(file_path+'weekdays/february/Tuesdays_feb.csv')
tue_february['Date'] = pd.to_datetime(tue_february['Date'])
tue_february_DL = tue_february.loc[:,cols_to_keep]
tue_february_index = tue_february_DL.set_index("Date")
                                             
'''*****************************************************************************************'''

wed_february = pd.read_csv(file_path+'weekdays/february/Wednesdays_feb.csv')
wed_february['Date'] = pd.to_datetime(wed_february['Date'])
wed_february_DL = wed_february.loc[:,cols_to_keep]
wed_february_index = wed_february_DL.set_index("Date")

'''*****************************************************************************************'''

thu_february = pd.read_csv(file_path+'weekdays/february/Thursdays_feb.csv')
thu_february['Date'] = pd.to_datetime(thu_february['Date']) 
thu_february_DL = thu_february.loc[:,cols_to_keep]
thu_february_index = thu_february_DL.set_index("Date")

'''*****************************************************************************************'''

fri_february = pd.read_csv(file_path+'weekdays/february/Fridays_feb.csv') 
fri_february['Date'] = pd.to_datetime(fri_february['Date']) 
fri_february_DL = fri_february.loc[:,cols_to_keep]
fri_february_index = fri_february_DL.set_index("Date")

'''*****************************************************************************************'''

sat_february = pd.read_csv(file_path+'weekdays/february/Saturdays_feb.csv')
sat_february['Date'] = pd.to_datetime(sat_february['Date'])
sat_february_DL = sat_february.loc[:,cols_to_keep]
sat_february_index = sat_february_DL.set_index("Date")

'''*****************************************************************************************'''
  
sun_february = pd.read_csv(file_path+'weekdays/february/Sundays_feb.csv') 
sun_february['Date'] = pd.to_datetime(sun_february['Date'])
sun_february_DL = sun_february.loc[:,cols_to_keep]
sun_february_index = sun_february_DL.set_index("Date")
```

```python
#creating an array of cells with no duplicates (original dataset has duplicates meaning that the same cell gathers traffic data for every hour)
ecell_id = mon_february_index["ECELL_ID"].to_list()
array_of_cells = []
for i in range(len(ecell_id)):
        if not (ecell_id[i] in array_of_cells):
            array_of_cells.append(ecell_id[i])
```

***************
MEDIAN COMPUTATION
***************

```python
#creating a new dataframe with hourly DL_VOL medians for each Monday of the month and every cell\n",
df0_feb = pd.DataFrame()
df1_feb = pd.DataFrame()
df2_feb = pd.DataFrame()
df3_feb = pd.DataFrame()
df4_feb = pd.DataFrame()
df5_feb = pd.DataFrame()
df6_feb = pd.DataFrame()

rows0 = []
rows1 = []
rows2 = []
rows3 = []
rows4 = []
rows5 = []
rows6 = []
   

for i in array_of_cells:
        
        df_idfiltered0 = mon_february_index[mon_february_index['ECELL_ID'] == i]
        df_idfiltered1 = tue_february_index[tue_february_index['ECELL_ID'] == i]
        df_idfiltered2 = wed_february_index[wed_february_index['ECELL_ID'] == i]
        df_idfiltered3 = thu_february_index[thu_february_index['ECELL_ID'] == i]
        df_idfiltered4 = fri_february_index[fri_february_index['ECELL_ID'] == i]
        df_idfiltered5 = sat_february_index[sat_february_index['ECELL_ID'] == i]
        df_idfiltered6 = sun_february_index[sun_february_index['ECELL_ID'] == i]
        
        x = 0
        while (x <= 23):
            df_hfiltered0 = df_idfiltered0[df_idfiltered0.index.hour == x]
            df_hfiltered1 = df_idfiltered1[df_idfiltered1.index.hour == x]
            df_hfiltered2 = df_idfiltered2[df_idfiltered2.index.hour == x]
            df_hfiltered3 = df_idfiltered3[df_idfiltered3.index.hour == x]
            df_hfiltered4 = df_idfiltered4[df_idfiltered4.index.hour == x]
            df_hfiltered5 = df_idfiltered5[df_idfiltered5.index.hour == x]
            df_hfiltered6 = df_idfiltered6[df_idfiltered6.index.hour == x]
            
            y = str(x)
            
            medianDL_VOL0 = df_hfiltered0['DL_VOL'].median()
            medianDL_VOL1 = df_hfiltered1['DL_VOL'].median()
            medianDL_VOL2 = df_hfiltered2['DL_VOL'].median()
            medianDL_VOL3 = df_hfiltered3['DL_VOL'].median()
            medianDL_VOL4 = df_hfiltered4['DL_VOL'].median()
            medianDL_VOL5 = df_hfiltered5['DL_VOL'].median()
            medianDL_VOL6 = df_hfiltered6['DL_VOL'].median()
            
            medianUL_VOL0 = df_hfiltered0['UL_VOL'].median()
            medianUL_VOL1 = df_hfiltered1['UL_VOL'].median()
            medianUL_VOL2 = df_hfiltered2['UL_VOL'].median()
            medianUL_VOL3 = df_hfiltered3['UL_VOL'].median()
            medianUL_VOL4 = df_hfiltered4['UL_VOL'].median()
            medianUL_VOL5 = df_hfiltered5['UL_VOL'].median()
            medianUL_VOL6 = df_hfiltered6['UL_VOL'].median()
            
            medianCS_SR0 = df_hfiltered0['CS_SR'].median()
            medianCS_SR1 = df_hfiltered1['CS_SR'].median()
            medianCS_SR2 = df_hfiltered2['CS_SR'].median()
            medianCS_SR3 = df_hfiltered3['CS_SR'].median()
            medianCS_SR4 = df_hfiltered4['CS_SR'].median()
            medianCS_SR5 = df_hfiltered5['CS_SR'].median()
            medianCS_SR6 = df_hfiltered6['CS_SR'].median()
            
            medianInterF_Hout_SR0 = df_hfiltered0['InterF_Hout_SR'].median()
            medianInterF_Hout_SR1 = df_hfiltered1['InterF_Hout_SR'].median()
            medianInterF_Hout_SR2 = df_hfiltered2['InterF_Hout_SR'].median()
            medianInterF_Hout_SR3 = df_hfiltered3['InterF_Hout_SR'].median()
            medianInterF_Hout_SR4 = df_hfiltered4['InterF_Hout_SR'].median()
            medianInterF_Hout_SR5 = df_hfiltered5['InterF_Hout_SR'].median()
            medianInterF_Hout_SR6 = df_hfiltered6['InterF_Hout_SR'].median()           
            
            
            rows0.append([i,"Monday at "+y,medianDL_VOL0,medianUL_VOL0,
                             medianCS_SR0,medianInterF_Hout_SR0])
            rows1.append([i,"Tuesday at "+y,medianDL_VOL1,medianUL_VOL1,
                             medianCS_SR1,medianInterF_Hout_SR1])
            rows2.append([i,"Wednesday at "+y,medianDL_VOL2,medianUL_VOL2,
                             medianCS_SR2,medianInterF_Hout_SR2])
            rows3.append([i,"Thursday at "+y,medianDL_VOL3,medianUL_VOL3,
                             medianCS_SR3,medianInterF_Hout_SR3])
            rows4.append([i,"Friday at "+y,medianDL_VOL4,medianUL_VOL4,
                             medianCS_SR4,medianInterF_Hout_SR4])
            rows5.append([i,"Saturday at "+y,medianDL_VOL5,medianUL_VOL5,
                             medianCS_SR5,medianInterF_Hout_SR5])
            rows6.append([i,"Sunday at "+y,medianDL_VOL6,medianUL_VOL6,
                             medianCS_SR6,medianInterF_Hout_SR6])

            x = x + 1

        df0_feb = pd.DataFrame(rows0)
        df1_feb = pd.DataFrame(rows1)
        df2_feb = pd.DataFrame(rows2)
        df3_feb = pd.DataFrame(rows3)
        df4_feb = pd.DataFrame(rows4)
        df5_feb = pd.DataFrame(rows5)
        df6_feb = pd.DataFrame(rows6)
    
df0_feb.columns = ["ECELL_ID","Hours","Median DL_VOL","Median UL_VOL", "Median CS_SR", "Median InterF_Hout_SR"]
df1_feb.columns = ["ECELL_ID","Hours","Median DL_VOL","Median UL_VOL", "Median CS_SR", "Median InterF_Hout_SR"]
df2_feb.columns = ["ECELL_ID","Hours","Median DL_VOL","Median UL_VOL", "Median CS_SR", "Median InterF_Hout_SR"]
df3_feb.columns = ["ECELL_ID","Hours","Median DL_VOL","Median UL_VOL", "Median CS_SR", "Median InterF_Hout_SR"]
df4_feb.columns = ["ECELL_ID","Hours","Median DL_VOL","Median UL_VOL", "Median CS_SR", "Median InterF_Hout_SR"]
df5_feb.columns = ["ECELL_ID","Hours","Median DL_VOL","Median UL_VOL", "Median CS_SR", "Median InterF_Hout_SR"]
df6_feb.columns = ["ECELL_ID","Hours","Median DL_VOL","Median UL_VOL", "Median CS_SR", "Median InterF_Hout_SR"]
df0_feb
    
```

```python
allcell_MWS_feb = pd.DataFrame()
for i in array_of_cells:
    cell1_mon = df0_feb[df0_feb['ECELL_ID'] == i]
    cell1_tue = df1_feb[df1_feb['ECELL_ID'] == i]
    cell1_wed = df2_feb[df2_feb['ECELL_ID'] == i]
    cell1_thu = df3_feb[df3_feb['ECELL_ID'] == i]
    cell1_fri = df4_feb[df4_feb['ECELL_ID'] == i]
    cell1_sat = df5_feb[df5_feb['ECELL_ID'] == i]
    cell1_sun = df6_feb[df6_feb['ECELL_ID'] == i]
    
    cell_MWS_feb = cell1_mon.append([cell1_tue, cell1_wed, cell1_thu, cell1_fri, cell1_sat, cell1_sun])
    allcell_MWS_feb=allcell_MWS_feb.append(cell_MWS_feb)
    
#cell_MWS.reset_index(drop=True, inplace=True)
allcell_MWS_feb.reset_index(drop=True, inplace=True)
allcell_MWS_feb
```

```python
#splitting allcell_MWS into dataframes by ECELL_IDs
dataframe_collection_feb = {} 
for i in array_of_cells:
    dataframe_collection_feb[i] = pd.DataFrame(allcell_MWS_feb[allcell_MWS_feb['ECELL_ID'] == i], columns=["ECELL_ID","Hours","Median DL_VOL","Median UL_VOL", "Median CS_SR", "Median InterF_Hout_SR"])
    
#we need the index to reset every 168 rows
    for j in dataframe_collection_feb:
        dataframe_collection_feb[j].reset_index(drop=True, inplace=True) 
dataframe_collection_feb[i]
#len(dataframe_collection) #this checks we have 395 dataframes (correct!)
```
