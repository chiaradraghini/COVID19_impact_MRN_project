```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from matplotlib.colors import ListedColormap
from datetime import datetime
import time
```

***************
DATA PROCESSING PHASE 1
***************

```python
file_path = '/Users/chiaradraghini/Documents/mrn_project/data/'

data_january = pd.read_csv(file_path+'Milano_800_January_New_MRN.csv') # Here January is selected
data_january['Date'] = pd.to_datetime(data_january['Date']) 
```

```python
Mondays = pd.DataFrame()
Tuesdays = pd.DataFrame()
Wednesdays = pd.DataFrame()
Thursdays = pd.DataFrame()
Fridays = pd.DataFrame()
Saturdays = pd.DataFrame()
Sundays = pd.DataFrame()

#first days of the week in January as starting point
Mon = 6
Tue = 7
Wed = 1
Thu = 2
Fri = 3
Sat = 4
Sun = 5

#creating new csv files for each day of the week with the corresponding hourly data 
while(Mon<28):
    monday = data_january[data_january['Date'] < pd.Timestamp(year=2020, month=1, day=(Mon+1), hour=0, minute=0)]
    monday = monday[monday['Date'] > pd.Timestamp(year=2020, month=1, day=(Mon-1), hour=23, minute=59)]
    Mon = Mon + 7
    Mondays = Mondays.append(monday)
Mondays.to_csv(r'/Users/chiaradraghini/Documents/mrn_project/data/weekdays/january/Mondays_jan.csv', index=False)

while(Tue<29):
    tuesday = data_january[data_january['Date'] < pd.Timestamp(year=2020, month=1, day=(Tue+1), hour=0, minute=0)]
    tuesday = tuesday[tuesday['Date'] > pd.Timestamp(year=2020, month=1, day=(Tue-1), hour=23, minute=59)]
    Tue = Tue + 7
    Tuesdays = Tuesdays.append(tuesday)
Tuesdays.to_csv(r'/Users/chiaradraghini/Documents/mrn_project/data/weekdays/january/Tuesdays_jan.csv', index=False)

while(Wed<30):
    wednesday = data_january[data_january['Date'] < pd.Timestamp(year=2020, month=1, day=(Wed+1), hour=0, minute=0)]
    wednesday = wednesday[wednesday['Date'] >= pd.Timestamp(year=2020, month=1, day=(Wed), hour=0, minute=0)]
    Wed = Wed + 7
    Wednesdays = Wednesdays.append(wednesday)
Wednesdays.to_csv(r'/Users/chiaradraghini/Documents/mrn_project/data/weekdays/january/Wednesdays_jan.csv', index=False)

while(Thu<31):
    thursday = data_january[data_january['Date'] < pd.Timestamp(year=2020, month=1, day=(Thu+1), hour=0, minute=0)]
    thursday = thursday[thursday['Date'] > pd.Timestamp(year=2020, month=1, day=(Thu-1), hour=23, minute=59)]
    Thu = Thu + 7
    Thursdays = Thursdays.append(thursday)
Thursdays.to_csv(r'/Users/chiaradraghini/Documents/mrn_project/data/weekdays/january/Thursdays_jan.csv', index=False)

while(Fri<32):
    friday = data_january[data_january['Date'] < pd.Timestamp(year=2020, month=1, day=(Fri), hour=23, minute=59)]
    friday = friday[friday['Date'] > pd.Timestamp(year=2020, month=1, day=(Fri-1), hour=23, minute=59)]
    Fri = Fri + 7
    Fridays = Fridays.append(friday)
Fridays.to_csv(r'/Users/chiaradraghini/Documents/mrn_project/data/weekdays/january/Fridays_jan.csv', index=False)

while(Sat<27):
    saturday = data_january[data_january['Date'] < pd.Timestamp(year=2020, month=1, day=(Sat+1), hour=0, minute=0)]
    saturday = saturday[saturday['Date'] > pd.Timestamp(year=2020, month=1, day=(Sat-1), hour=23, minute=59)]
    Sat = Sat + 7
    Saturdays = Saturdays.append(saturday)
Saturdays.to_csv(r'/Users/chiaradraghini/Documents/mrn_project/data/weekdays/january/Saturdays_jan.csv', index=False)

while(Sun<32):
    sunday = data_january[data_january['Date'] < pd.Timestamp(year=2020, month=1, day=(Sun+1), hour=0, minute=0)]
    sunday = sunday[sunday['Date'] > pd.Timestamp(year=2020, month=1, day=(Sun-1), hour=23, minute=59)]
    Sun = Sun + 7
    Sundays = Sundays.append(sunday)
Sundays.to_csv(r'/Users/chiaradraghini/Documents/mrn_project/data/weekdays/january/Sundays_jan.csv', index=False)
```

***************
DATA PROCESSING PHASE 2
***************

```python
cols_to_keep = ['ECELL_ID', 'Date', 'DL_VOL', 'UL_VOL', 'CS_SR', 'InterF_Hout_SR']

mon_january = pd.read_csv(file_path+'weekdays/january/Mondays_jan.csv')
mon_january['Date'] = pd.to_datetime(mon_january['Date'])
mon_january_DL = mon_january.loc[:,cols_to_keep]
mon_january_index = mon_january_DL.set_index("Date")

'''*****************************************************************************************'''
                                             
tue_january = pd.read_csv(file_path+'weekdays/january/Tuesdays_jan.csv')
tue_january['Date'] = pd.to_datetime(tue_january['Date'])
tue_january_DL = tue_january.loc[:,cols_to_keep]
tue_january_index = tue_january_DL.set_index("Date")
                                             
'''*****************************************************************************************'''

wed_january = pd.read_csv(file_path+'weekdays/january/Wednesdays_jan.csv')
wed_january['Date'] = pd.to_datetime(wed_january['Date'])
wed_january_DL = wed_january.loc[:,cols_to_keep]
wed_january_index = wed_january_DL.set_index("Date")

'''*****************************************************************************************'''

thu_january = pd.read_csv(file_path+'weekdays/january/Thursdays_jan.csv')
thu_january['Date'] = pd.to_datetime(thu_january['Date']) 
thu_january_DL = thu_january.loc[:,cols_to_keep]
thu_january_index = thu_january_DL.set_index("Date")

'''*****************************************************************************************'''

fri_january = pd.read_csv(file_path+'weekdays/january/Fridays_jan.csv') 
fri_january['Date'] = pd.to_datetime(fri_january['Date']) 
fri_january_DL = fri_january.loc[:,cols_to_keep]
fri_january_index = fri_january_DL.set_index("Date")

'''*****************************************************************************************'''

sat_january = pd.read_csv(file_path+'weekdays/january/Saturdays_jan.csv')
sat_january['Date'] = pd.to_datetime(sat_january['Date'])
sat_january_DL = sat_january.loc[:,cols_to_keep]
sat_january_index = sat_january_DL.set_index("Date")

'''*****************************************************************************************'''
  
sun_january = pd.read_csv(file_path+'weekdays/january/Sundays_jan.csv') 
sun_january['Date'] = pd.to_datetime(sun_january['Date'])
sun_january_DL = sun_january.loc[:,cols_to_keep]
sun_january_index = sun_january_DL.set_index("Date")
```

```python
#creating an array of cells with no duplicates (original dataset has duplicates meaning that the same cell gathers traffic data for every hour)
ecell_id = mon_january_index["ECELL_ID"].to_list()
array_of_cells = []
for i in range(len(ecell_id)):
        if not (ecell_id[i] in array_of_cells):
            array_of_cells.append(ecell_id[i])
%store array_of_cells 
```

***************
MEDIAN COMPUTATION
***************

```python
#creating a new dataframe with hourly DL_VOL medians for each Monday of the month and every cell
df0_jan = pd.DataFrame()
df1_jan = pd.DataFrame()
df2_jan = pd.DataFrame()
df3_jan = pd.DataFrame()
df4_jan = pd.DataFrame()
df5_jan = pd.DataFrame()
df6_jan = pd.DataFrame()

rows0 = []
rows1 = []
rows2 = []
rows3 = []
rows4 = []
rows5 = []
rows6 = []
   

for i in array_of_cells:
        
        df_idfiltered0 = mon_january_index[mon_january_index['ECELL_ID'] == i]
        df_idfiltered1 = tue_january_index[tue_january_index['ECELL_ID'] == i]
        df_idfiltered2 = wed_january_index[wed_january_index['ECELL_ID'] == i]
        df_idfiltered3 = thu_january_index[thu_january_index['ECELL_ID'] == i]
        df_idfiltered4 = fri_january_index[fri_january_index['ECELL_ID'] == i]
        df_idfiltered5 = sat_january_index[sat_january_index['ECELL_ID'] == i]
        df_idfiltered6 = sun_january_index[sun_january_index['ECELL_ID'] == i]
        
        x = 0
        while (x <= 23):
            df_hfiltered0 = df_idfiltered0[df_idfiltered0.index.hour == x]
            df_hfiltered1 = df_idfiltered1[df_idfiltered1.index.hour == x]
            df_hfiltered2 = df_idfiltered2[df_idfiltered2.index.hour == x]
            df_hfiltered3 = df_idfiltered3[df_idfiltered3.index.hour == x]
            df_hfiltered4 = df_idfiltered4[df_idfiltered4.index.hour == x]
            df_hfiltered5 = df_idfiltered5[df_idfiltered5.index.hour == x]
            df_hfiltered6 = df_idfiltered6[df_idfiltered6.index.hour == x]
            
            y = str(x)
            
            medianDL_VOL0 = df_hfiltered0['DL_VOL'].median()
            medianDL_VOL1 = df_hfiltered1['DL_VOL'].median()
            medianDL_VOL2 = df_hfiltered2['DL_VOL'].median()
            medianDL_VOL3 = df_hfiltered3['DL_VOL'].median()
            medianDL_VOL4 = df_hfiltered4['DL_VOL'].median()
            medianDL_VOL5 = df_hfiltered5['DL_VOL'].median()
            medianDL_VOL6 = df_hfiltered6['DL_VOL'].median()
            
            medianUL_VOL0 = df_hfiltered0['UL_VOL'].median()
            medianUL_VOL1 = df_hfiltered1['UL_VOL'].median()
            medianUL_VOL2 = df_hfiltered2['UL_VOL'].median()
            medianUL_VOL3 = df_hfiltered3['UL_VOL'].median()
            medianUL_VOL4 = df_hfiltered4['UL_VOL'].median()
            medianUL_VOL5 = df_hfiltered5['UL_VOL'].median()
            medianUL_VOL6 = df_hfiltered6['UL_VOL'].median()
            
            medianCS_SR0 = df_hfiltered0['CS_SR'].median()
            medianCS_SR1 = df_hfiltered1['CS_SR'].median()
            medianCS_SR2 = df_hfiltered2['CS_SR'].median()
            medianCS_SR3 = df_hfiltered3['CS_SR'].median()
            medianCS_SR4 = df_hfiltered4['CS_SR'].median()
            medianCS_SR5 = df_hfiltered5['CS_SR'].median()
            medianCS_SR6 = df_hfiltered6['CS_SR'].median()
            
            medianInterF_Hout_SR0 = df_hfiltered0['InterF_Hout_SR'].median()
            medianInterF_Hout_SR1 = df_hfiltered1['InterF_Hout_SR'].median()
            medianInterF_Hout_SR2 = df_hfiltered2['InterF_Hout_SR'].median()
            medianInterF_Hout_SR3 = df_hfiltered3['InterF_Hout_SR'].median()
            medianInterF_Hout_SR4 = df_hfiltered4['InterF_Hout_SR'].median()
            medianInterF_Hout_SR5 = df_hfiltered5['InterF_Hout_SR'].median()
            medianInterF_Hout_SR6 = df_hfiltered6['InterF_Hout_SR'].median()           
            
            
            rows0.append([i,"Monday at "+y,medianDL_VOL0,medianUL_VOL0,
                             medianCS_SR0,medianInterF_Hout_SR0])
            rows1.append([i,"Tuesday at "+y,medianDL_VOL1,medianUL_VOL1,
                             medianCS_SR1,medianInterF_Hout_SR1])
            rows2.append([i,"Wednesday at "+y,medianDL_VOL2,medianUL_VOL2,
                             medianCS_SR2,medianInterF_Hout_SR2])
            rows3.append([i,"Thursday at "+y,medianDL_VOL3,medianUL_VOL3,
                             medianCS_SR3,medianInterF_Hout_SR3])
            rows4.append([i,"Friday at "+y,medianDL_VOL4,medianUL_VOL4,
                             medianCS_SR4,medianInterF_Hout_SR4])
            rows5.append([i,"Saturday at "+y,medianDL_VOL5,medianUL_VOL5,
                             medianCS_SR5,medianInterF_Hout_SR5])
            rows6.append([i,"Sunday at "+y,medianDL_VOL6,medianUL_VOL6,
                             medianCS_SR6,medianInterF_Hout_SR6])
            
            #print("Median of hour",x,"and ecell id",i,":",median)
            x = x + 1

        df0_jan = pd.DataFrame(rows0)
        df1_jan = pd.DataFrame(rows1)
        df2_jan = pd.DataFrame(rows2)
        df3_jan = pd.DataFrame(rows3)
        df4_jan = pd.DataFrame(rows4)
        df5_jan = pd.DataFrame(rows5)
        df6_jan = pd.DataFrame(rows6)
    
df0_jan.columns = ["ECELL_ID","Hours","Median DL_VOL","Median UL_VOL", "Median CS_SR", "Median InterF_Hout_SR"]
df1_jan.columns = ["ECELL_ID","Hours","Median DL_VOL","Median UL_VOL", "Median CS_SR", "Median InterF_Hout_SR"]
df2_jan.columns = ["ECELL_ID","Hours","Median DL_VOL","Median UL_VOL", "Median CS_SR", "Median InterF_Hout_SR"]
df3_jan.columns = ["ECELL_ID","Hours","Median DL_VOL","Median UL_VOL", "Median CS_SR", "Median InterF_Hout_SR"]
df4_jan.columns = ["ECELL_ID","Hours","Median DL_VOL","Median UL_VOL", "Median CS_SR", "Median InterF_Hout_SR"]
df5_jan.columns = ["ECELL_ID","Hours","Median DL_VOL","Median UL_VOL", "Median CS_SR", "Median InterF_Hout_SR"]
df6_jan.columns = ["ECELL_ID","Hours","Median DL_VOL","Median UL_VOL", "Median CS_SR", "Median InterF_Hout_SR"]
df0_jan
```

```python
#storing for later use in the creation of kmeans dataset
%store df0_jan
%store df1_jan
%store df2_jan
%store df3_jan 
%store df4_jan
%store df5_jan
%store df6_jan      
```

***********
BUILDING CELL DATAFRAME

```python
allcell_MWS_jan = pd.DataFrame()
for i in array_of_cells:
    cell1_mon = df0_jan[df0_jan['ECELL_ID'] == i]
    cell1_tue = df1_jan[df1_jan['ECELL_ID'] == i]
    cell1_wed = df2_jan[df2_jan['ECELL_ID'] == i]
    cell1_thu = df3_jan[df3_jan['ECELL_ID'] == i]
    cell1_fri = df4_jan[df4_jan['ECELL_ID'] == i]
    cell1_sat = df5_jan[df5_jan['ECELL_ID'] == i]
    cell1_sun = df6_jan[df6_jan['ECELL_ID'] == i]
    
    cell_MWS_jan = cell1_mon.append([cell1_tue, cell1_wed, cell1_thu, cell1_fri, cell1_sat, cell1_sun])
    allcell_MWS_jan=allcell_MWS_jan.append(cell_MWS_jan)
    
allcell_MWS_jan.reset_index(drop=True, inplace=True)
allcell_MWS_jan
```

```python
#splitting allcell_MWS into dataframes by ECELL_IDs
dataframe_collection_jan = {} 
for i in array_of_cells:
    dataframe_collection_jan[i] = pd.DataFrame(allcell_MWS_jan[allcell_MWS_jan['ECELL_ID'] == i], columns=["ECELL_ID","Hours","Median DL_VOL","Median UL_VOL", "Median CS_SR", "Median InterF_Hout_SR"])
    #print(dataframe_collection_jan[i])
    
#we need the index to reset every 168 rows
    for j in dataframe_collection_jan:
        dataframe_collection_jan[j].reset_index(drop=True, inplace=True) 
dataframe_collection_jan[i]
#len(dataframe_collection_jan) #this checks we have 395 dataframes (correct!)
%store dataframe_collection_jan
dataframe_collection_jan[i]
```
